**Correlation Matrix**

Correlation is used to measure the extent to which one variable changes with respect to another variable.

```python
corr_matrix = df.corr()
```

```python
heatmap = sns.heatmap(corr_matrix, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
```

The white represents that the correlation value is 1, which means those two variables are the same. The black color represents the correlation value is -1, which means the two variables are completely dependent but in negative directions. One particular color between black and white represents the correlation value of 0, which means two variables are completely independent. 

The features with a higher absolute value of correlation are beneficial. Also we can remove the features with no correlation.

```python
df1 = pd.read_csv("powerconsumption.csv", parse_dates = True, index_col = "Datetime")
```

```python
# Resampling the time series data based on monthly 'M' frequency
df_month = df1.resample("M").mean()

# using subplot
fig, ax = plt.subplots(figsize=(10, 6))

# plotting bar graph
ax.bar(df_month['2017':].index,
       df_month.loc['2017':, "PowerConsumption_Zone3"],
       width=25, align='center')
```
